```python
import tensorflow as tf

# Load the trained model
model = tf.keras.models.load_model('/content/drive/MyDrive/children and adults/childandadults.h5')
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Load and preprocess the image you want to classify
img_path = '/content/drive/MyDrive/children and adults/test/children/18.jpg'  # Replace with the path to your image
img = image.load_img(img_path, target_size=(150, 150))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0  # Normalize the image

# Make the prediction
prediction = model.predict(img_array)

# Define labels and colors based on the prediction
if prediction >= 0.5:
    label = 'Adults'
    border_color = 'green'
else:
    label = 'Children'
    border_color = 'red'

# Display the image with label and colored border
plt.imshow(img)
plt.title(label, fontsize=14, color='black', backgroundcolor=border_color)
plt.axis('off')
plt.show()
```
